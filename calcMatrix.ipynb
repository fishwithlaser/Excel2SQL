```python
# CalcMatrix
# construct linear metamorphic reactions with the input phases.
# =======================================
# phase is the data-structure for solid solution models.
# A0 is the composition matrix before Gaussian elimination
# C is the Modal composition

def calcMatrix():
    
#### I think this looks at the number of phases that I'm going to calc?
#
#p=length(phase);            % number of phases
#l=1;
#for i=1:p
#    for j=1:phase{i}.endnum
#        A(l,:)=dataset(phase{i}.endmember(j)).Comp;      % Compositional matrix
#        l=l+1;
#    end    
#end

#[i,l]=size(A);j=1;A_temp=zeros(i,1);
#for i=1:l
#    if max(A(:,i))>0
#        A_temp(:,j)=A(:,i);
#        C{j}=CC{i};
#        j=j+1;
#    end   
#end
#clear A;A=A_temp;clear A_temp;A=A';A0=A;
#[m,n]=size(A);

i=1;j=1;l=zeros(1,m);   %l vector is used to record the first non-zero element in i-th row

while i<=m
    
    % partial pivoting
    
    if j>n
        React=0;A0=A0*0;
        return;
    end
    
    p=abs(A(i,j));q=i;
    for k=(i+1):m
        if abs(A(k,j))>p
            p=abs(A(k,j));q=k;
        end
    end
    
    if abs(p)>1e-14
        A_temp=A(q,j:n);
        A(q,j:n)=A(i,j:n);
        A(i,j:n)=A_temp/A_temp(1);
    
    %Gaussian Elimination
        for k=(i+1):m
            A(k,:)=(A(k,:)-A(i,:)*A(k,j));
        end
        for k=1:i-1
            A(k,:)=(A(k,:)-A(i,:)*A(k,j));
        end
        l(i)=j;j=j+1;i=i+1;
    else
        A(i:m,j)=0;
        l(i)=j;j=j+1;
    end
end
l(m+1)=n+1;

% eliminate round-off errors in the matrix
for i=1:m
    for j=i+1:n
        if abs(A(i,j))<1e-14
            A(i,j)=0;
        end
    end
end

% independent metamorphic reactions
React=zeros(n,n-m);             % m-n independent metamorphic reactions
A1=zeros(m,m);A2=zeros(m,n-m);
m1=zeros(1,m);m2=zeros(1,n-m);

j=1;k=1;
for i=1:m
    while j<l(i+1)
        if j==l(i)
            A1(:,i)=A(:,j);
            m1(i)=j;
            j=j+1;
        else
            A2(:,k)=A(:,j);
            m2(k)=j;
            React(j,k)=1;
            k=k+1;
            j=j+1;
        end
    end
end

b=eye(n-m);
for i=1:n-m
    React(m1,i)=-A2*b(:,i);
end

end
```
